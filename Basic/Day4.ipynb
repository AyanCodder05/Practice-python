```python
#1. Create a DataFrame from a NumPy array and plot its columns as a simple line plot using Matplotlib.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])
df=pd.DataFrame({'X':x,'Y':y})
print(df)
plt.plot(df['X'],df['Y'],marker='o')
plt.show()
```

```python
#2. Convert a Series to a NumPy array, compute its histogram bins with np.histogram(), and draw the histogram with Matplotlib.
s = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 5])
arr=s.to_numpy(s)
print(arr)
hist,bins=np.histogram(arr, bins=5)#np.histogram() divides data into equal-sized bins and counts how many values fall into each bin.
print("Histogram:",hist)
print("Bins", bins)

plt.hist(arr, bins=5, color='skyblue', edgecolor='black')
plt.show()
```

```python
#3. Create a DataFrame with np.random.randn() and show a Seaborn pairplot.
import seaborn as sns
data=np.random.randn(100,4)
print("Data", data)

df=pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])
print("Data Frame:\n", df)

sns.pairplot(df)
plt.show()
```

```python
#4. Make a (55) NumPy array  DataFrame with columns AE; plot a heatmap of the array with Seaborn.
data=np.arange(1,26).reshape(5,5)
print("Data:",data)

df=pd.DataFrame(data, columns=['A','B','C','D','E'])
print("Data Frame:\n",df)

sns.heatmap(df)
plt.show()
```

```python
#5. Create a DataFrame with 10 sequential numbers using np.arange() and plot a stemplot (Matplotlib).
data=np.arange(1,10)
print("Data:", data)

df=pd.DataFrame(data,columns=['Numbers'])
print("Data Frame",df)


plt.stem(df.index, df['Numbers'])
plt.title("Stemplot of Sequential Numbers")
plt.xlabel("Index")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
